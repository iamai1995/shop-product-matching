```python
import numpy as np
import pandas as pd
import random as random
import matplotlib.pyplot as plt
```

# Generate random indices

```python
# generate a list of random indexes
random.seed(1)
random_index = random.sample(range(34250), 500)
len(random_index)
```

```python
# check for duplicated indices
len(set(random_index))
```

```python
plt.hist(random_index)
plt.show()
```
